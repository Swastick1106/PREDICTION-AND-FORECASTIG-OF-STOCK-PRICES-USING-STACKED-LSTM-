```python
data = pd.read_csv('AAPL.csv')
data.head()
```

```python
data1 = data.reset_index()['close']
data1.shape
```

```python
plt.plot(data1)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
data1 = scaler.fit_transform(np.array(data1).reshape(-1,1))
```

```python
print(data1)
```

```python
training_size = int(len(data1)*0.70)   #70% of total length of the dataset
test_size = len(data1) - training_size
print(training_size)
print(test_size)

train_data,test_data = data1[0:training_size,:], data1[training_size:len(data1),:1]
print(train_data)
print(test_data)
```

Considering Time Stamps - basically it is the bracket of the number of days which will be considerd to next output 

This helpes in capturing the patterns and minute details in the data.

```python
def create_Dataset(dataset,time_step=1):
    dataX, dataY =[],[]
    for i in range (len(dataset)-time_step-1):
        a = dataset[i:(i+time_step),0]
        dataX.append(a)
        dataY.append(dataset[i+time_step, 0])
    return np.array(dataX), np.array(dataY)
```

```python
time_step = 100
X_train,y_train = create_Dataset(train_data,time_step)
X_test, y_test = create_Dataset(test_data,time_step)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

```python
X_train = X_train.reshape(X_train.shape[0],X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

#### LSTM Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model = Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences = True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=100, batch_size=32, verbose=1)
```

```python
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
```

```python
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(y_train,train_predict))
```

```python
sns.set(style="whitegrid", context="talk", palette="deep")

# Define look_back
look_back = 100

# Create placeholder arrays with NaNs
trainPredictPlot = np.empty_like(data1)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(train_predict)+look_back, :] = train_predict

testPredictPlot = np.empty_like(data1)
testPredictPlot[:, :] = np.nan
testPredictPlot[len(train_predict)+(look_back*2)+1:len(data1)-1, :] = test_predict

# Create a time axis
time_axis = np.arange(len(data1))

# Plot
plt.figure(figsize=(14, 6))
plt.plot(time_axis, scaler.inverse_transform(data1), label='Actual Stock Price', color=sns.color_palette()[0])
plt.plot(time_axis, trainPredictPlot, label='Training Predictions', color=sns.color_palette("muted")[2])
plt.plot(time_axis, testPredictPlot, label='Test Predictions', color=sns.color_palette("muted")[3])

plt.title("Stock Price: Prediction vs Actual")
plt.xlabel("Time Steps")
plt.ylabel("Price")
plt.legend()
plt.tight_layout()
plt.show()
```
